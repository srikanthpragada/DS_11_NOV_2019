### apply() and applymap()

```python
def get_per_tip(row):
    return  round(row['tip'] / row['total_bill']  * 100)
```

```python
def non_null_count(s):
    return  s.isna().sum()
```

```python
df.apply(lambda s : s.isna().sum())
```

```python
df.apply(non_null_count)
```

```python
df.apply(get_per_tip,axis=1)
```

```python
def get_tip_type(s):
    per = s['tip'] / s['total_bill'] * 100
    if per < 10:
        return 'Low'
    elif per < 20:
        return 'Medium'
    else:
        return 'High'
        
df.apply(get_tip_type, axis=1)
```

```python
def encode_week(v):
    weekdict = {'Sun':1, 'Sat' : 7, 'Fri':6, 'Thur' : 5}
    return  weekdict.get(v,0)
```

```python
df['day'].apply(encode_week)
```

```python
df.loc[0:10,['total_bill','tip']].apply(lambda s : s[s.notna()].count(), axis=1)
```

```python
df.loc[0:4,['day','time']].applymap(lambda v : v.upper())
```

```python
df.loc[0:4,['total_bill','tip']].applymap(lambda v : ("%5.0f" % v))
```
