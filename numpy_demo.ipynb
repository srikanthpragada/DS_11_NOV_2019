### Broadcasting

```python
a = np.array([1,2,3,4,5])
```

```python
a + 5
```

```python
a1 = np.array ([1,2,3])
a2 = np.array ([[1,2,3],[4,5,6]])
a1 + a2
```

```python
# Error as no. of cols is not same 
a1 = np.array ([1,2])   # 2 cols
a2 = np.array ([[1,2,3],[4,5,6]])   # 3 cols 
a1 + a2
```

```python
# Error in broadcasting due to incompatible shapes 
a1 = np.ones((3,3))
print(a1)
a2 = np.ones((2,3))
print(a2)
a1 + a2
```
